### Metodi di Esportazione

```python
df1.to_csv('df1.csv', columns=['column_1', 'column_5', 'column_2'], index_label='ID')
```

```python
df1.to_pickle('df1.pkl')
```

Meglio l'utilizzo di pickle se la tabella contiene moltissimi numeri, dato che csv salva tutto come stringhe, mentre con pickle si pu ottimizzare la memoria

### Metodi di Visualizzazione

```python
df1_plot = df1.sort_values('column_1')
df1_plot.plot('column_1', ['column_6', 'column_3'], figsize=(12,8))
```

```python
df1.plot.scatter(x='column_2', y='column_5', c='column_3', colormap='autumn', figsize=(12,8))
```

```python
df1.boxplot(column=['column_3', 'column_2'], figsize=(12,8))
```

```python
df1.loc[:, df1.columns!='column_1'].plot.bar(figsize=(12,8))
```

```python
df1.loc[:, df1.columns!='column_1'].plot.density(figsize=(12,8))
```
